Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
def agegrp(age):
  if age<30:
    return "Youth"
  elif age >= 30 and age<=60:
    return "Adult"
  else:
    return "Senior"

data.assign(age_group=data['age'].apply(agegrp))
```
